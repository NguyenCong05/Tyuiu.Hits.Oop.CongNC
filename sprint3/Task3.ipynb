```
using System;
using System.Collections.Generic;

// Базовый класс Animal
public class Animal
{
    public string Name { get; set; }
    public int Age { get; set; }
    public string Habitat { get; set; } // новый атрибут
    public bool IsDomestic { get; set; } // новый атрибут

    public Animal(string name, int age, string habitat, bool isDomestic)
    {
        Name = name;
        Age = age;
        Habitat = habitat;
        IsDomestic = isDomestic;
    }

    // Виртуальный метод, который можно переопределять
    public virtual void MakeSound()
    {
        Console.WriteLine($"{Name} издает общий звук.");
    }

    // Метод для отображения информации
    public virtual void ShowInfo()
    {
        Console.WriteLine($"Имя: {Name}, Возраст: {Age}, Место обитания: {Habitat}, Домашний: {IsDomestic}");
    }

    // Метод с перегрузкой: бегать
    public virtual void Run()
    {
        Console.WriteLine($"{Name} бежит.");
    }

    public void Run(int speed)
    {
        Console.WriteLine($"{Name} бежит со скоростью {speed} км/ч.");
    }
}

// Производный класс Dog
public class Dog : Animal
{
    public string Breed { get; set; }
    public bool IsTrained { get; set; } // новый атрибут
    public string FavoriteToy { get; set; } // новый атрибут

    public Dog(string name, int age, string habitat, bool isDomestic, string breed, bool isTrained, string favoriteToy)
        : base(name, age, habitat, isDomestic)
    {
        Breed = breed;
        IsTrained = isTrained;
        FavoriteToy = favoriteToy;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} говорит: Гав-гав!");
    }

    public void Fetch()
    {
        Console.WriteLine($"{Name} приносит мяч.");
    }

    // Перегрузка метода Run
    public override void Run()
    {
        Console.WriteLine($"{Name} бегает быстро!");
    }

    public new void Run(int speed)
    {
        Console.WriteLine($"{Name} бежит со скоростью {speed} км/ч.");
    }
}

// Производный класс Cat
public class Cat : Animal
{
    public string Color { get; set; }
    public bool IsHungry { get; set; } // новый атрибут
    public int LivesLeft { get; set; } // новый атрибут

    public Cat(string name, int age, string habitat, bool isDomestic, string color, bool isHungry, int livesLeft)
        : base(name, age, habitat, isDomestic)
    {
        Color = color;
        IsHungry = isHungry;
        LivesLeft = livesLeft;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} говорит: Мяу-мяу!");
    }

    public void Purr()
    {
        Console.WriteLine($"{Name} мурлычет.");
    }

    public override void Run()
    {
        Console.WriteLine($"{Name} быстро бегает по дому.");
    }
}

// Производный класс Bird
public class Bird : Animal
{
    public string Species { get; set; }
    public double WingSpan { get; set; } // Новый атрибут
    public bool CanFly { get; set; } // Новый атрибут

    public Bird(string name, int age, string habitat, bool isDomestic, string species, double wingSpan, bool canFly)
        : base(name, age, habitat, isDomestic)
    {
        Species = species;
        WingSpan = wingSpan;
        CanFly = canFly;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} поет: Чирик-чирик!");
    }

    public void Fly()
    {
        if (CanFly)
            Console.WriteLine($"{Name} летит в небо.");
        else
            Console.WriteLine($"{Name} не умеет летать.");
    }
}

// Универсальный класс для коллекции животных
public class AnimalCollection<T> where T : Animal
{
    private List<T> animals = new List<T>();

    public void Add(T animal)
    {
        animals.Add(animal);
    }

    public void ShowAll()
    {
        foreach (var animal in animals)
        {
            animal.ShowInfo();
            animal.MakeSound();

            // Вызов перегруженного метода Run
            animal.Run();
            animal.Run(20);
            Console.WriteLine();
        }
    }
}


    {
        // Создаем коллекции
        var dogs = new AnimalCollection<Dog>();
        dogs.Add(new Dog("Рэкс", 5, "Пасека", true, "Немецкая овчарка", true, "Мяч"));
        dogs.Add(new Dog("Бобик", 3, "Дача", true, "Мопс", false, "Кость"));

        var cats = new AnimalCollection<Cat>();
        cats.Add(new Cat("Мурка", 2, "Квартира", true, "Серый", true, 9));
        cats.Add(new Cat("Барсик", 4, "Дом", true, "Рыжий", false, 7));

        var birds = new AnimalCollection<Bird>();
        birds.Add(new Bird("Кеша", 1, "Город", true, "Попугай", 15.2, true));
        birds.Add(new Bird("Птица", 2, "Лес", false, "Ворона", 25.0, true));

        Console.WriteLine("Доги:");
        dogs.ShowAll();

        Console.WriteLine("Кошки:");
        cats.ShowAll();

        Console.WriteLine("Птицы:");
        birds.ShowAll();
    }
```
