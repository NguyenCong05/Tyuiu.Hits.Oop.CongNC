```
using System;
using System.Collections.Generic;

// --- Сервисы и интерфейсы ---
public interface INotificationService
{
    void Register(string message);
}


public interface IAnalyticsService
{
    void Register(string message);
    void TrackProductView(string productName);
    void TrackSale(string productName, int quantity);
 } 
// Реализация сервисов
public class NotificationService : INotificationService
{
    public void Register(string message)
    {
        Console.WriteLine("Сервис INotificationService зарегистрирован");
    }
}
   
public class AnalyticsService : IAnalyticsService
{
    public void Register(string message)
    {
        Console.WriteLine("Сервис IAnalyticsService зарегистрирован");
    }

    public void TrackProductView(string productName)
    {
        Console.WriteLine($"Аналитика: Просмотр товара '{productName}'");
    }

    public void TrackSale(string productName, int quantity)
    {
        Console.WriteLine($"Аналитика: Продано {quantity} шт. товара '{productName}'");
    }
}

// --- Базовые и расширенные классы товаров ---
public abstract class Product
{
    public string Name { get; set; }
    public decimal Price { get; set; }
    public int Quantity { get; set; }
    public string Producer { get; set; }

    public Product(string name, decimal price, int quantity, string producer)
    {
        Name = name;
        Price = price;
        Quantity = quantity;
        Producer = producer;
    }

    public abstract void ShowDetails();

    public void ChangePrice(decimal newPrice)
    {
        Console.WriteLine($"Цена товара '{Name}' изменена: {Price} -> {newPrice}");
        Price = newPrice;
    }
}

// Электроника
public class Electronics : Product
{
    public string Category { get; set; }
    public bool IsExpired { get; set; }
    public string Color { get; set; }
    public double Weight { get; set; }
    public int EnergyRating { get; set; } // класс энергоэффективности

    public Electronics(string name, decimal price, int quantity, string producer, string category,
        bool isExpired, string color, double weight, int energyRating)
        : base(name, price, quantity, producer)
    {
        Category = category;
        IsExpired = isExpired;
        Color = color;
        Weight = weight;
        EnergyRating = energyRating;
    }

    public override void ShowDetails()
    {
        Console.WriteLine($"Название: {Name}; Цена: {Price}; Производитель: {Producer}; Количество: {Quantity}");
        Console.WriteLine($"Категория: {Category}; Просрочен: {IsExpired}");
        Console.WriteLine($"Цвет: {Color}; Вес: {Weight}кг; Класс энергоэффективности: {EnergyRating}");
    }

    public void ConnectBluetooth()
    {
        Console.WriteLine($"Устройство '{Name}' подключено по Bluetooth");
    }
}

// Одежда
public class Clothing : Product
{
    public string Size { get; set; }
    public string Material { get; set; }
    public string Season { get; set; }
    public bool IsEcoFriendly { get; set; }

    public Clothing(string name, decimal price, int quantity, string producer, string size, string material, string season)
        : base(name, price, quantity, producer)
    {
        Size = size;
        Material = material;
        Season = season;
    }

    public override void ShowDetails()
    {
        Console.WriteLine($"Название: {Name}; Цена: {Price}; Производитель: {Producer}; Количество: {Quantity}");
        Console.WriteLine($"Размер: {Size}; В наличии, Материал: {Material}; Сезон: {Season}");
    }

    public void MarkEcoFriendly()
    {
        IsEcoFriendly = true;
        Console.WriteLine($"Одежда '{Name}' помечена как экологически чистая");
    }

    public void IncreaseQuantity(int amount)
    {
        Quantity += amount;
        Console.WriteLine($"Количество товара '{Name}' увеличено на {amount}. Теперь: {Quantity}");
    }
}

// Книги
public class Book : Product
{
    public string Author { get; set; }
    public int Pages { get; set; }
    public string Genre { get; set; }
    public string Publisher { get; set; }

    public Book(string name, decimal price, int quantity, string producer, string author, int pages, string genre, string publisher)
        : base(name, price, quantity, producer)
    {
        Author = author;
        Pages = pages;
        Genre = genre;
        Publisher = publisher;
    }

    public override void ShowDetails()
    {
        Console.WriteLine($"Книга: {Name}; Автор: {Author}; Страниц: {Pages}; Жанр: {Genre}; Издатель: {Publisher}; Цена: {Price}");
    }
}

// --- Менеджеры ---
public class StoreManager
{
    private readonly List<Product> _products = new List<Product>();
    private readonly INotificationService _notificationService;
    private readonly IAnalyticsService _analyticsService;

    public StoreManager(INotificationService notif, IAnalyticsService analytics)
    {
        _notificationService = notif;
        _analyticsService = analytics;
    }

    public void AddProduct(Product product)
    {
        _products.Add(product);
        Console.WriteLine($"Товар '{product.Name}' добавлен в менеджер");
        _analyticsService.TrackProductView(product.Name);
    }

    public void ShowAllProducts()
    {
        Console.WriteLine("=== ВСЕ ТОВАРЫ ===");
        foreach (var p in _products)
        {
            p.ShowDetails();
            Console.WriteLine();
        }
    }

    public void ChangeProductPrice(string productName, decimal newPrice)
    {
        var product = _products.Find(p => p.Name == productName);
        if (product != null)
        {
            decimal oldPrice = product.Price;
            product.ChangePrice(newPrice);
            _notificationService.Register($"Скидка на {productName}: {oldPrice} -> {newPrice}");
            _analyticsService.TrackSale(productName, 1);
        }
    }
}

// --- Основная программа ---

    {
        // Регистрация сервисов
        INotificationService notifService = new NotificationService();
        IAnalyticsService analyticsService = new AnalyticsService();
        notifService.Register("Регистрация сервисов");
        analyticsService.Register("Регистрация сервисов");

        // Создаем менеджер магазина
        var storeManager = new StoreManager(notifService, analyticsService);

        // Создаем и добавляем товары электроника
        var laptop = new Electronics("MagicBook X16 Plus", 64490m, 2, "Honor", "X16 Plus", false, "Серебристый", 1.8, 3);
        var smartphone = new Electronics("Смартфон", 29990m, 1, "Samsung", "Galaxy S21", false, "Черный", 0.2, 4);
        var tv = new Electronics("Телевизор Samsung", 45000m, 3, "Samsung", "QLED 55\"", false, "Черный", 8.5, 2);

        // Создаем и добавляем товары одежда
        var hoodie = new Clothing("худи Casual", 4599m, 13, "Befree", "50", "Хлопок", "Осень-Зима");
        var tShirt = new Clothing("Футболка", 999m, 2, "Nike", "48", "Хлопок", "Лето");
        var jacket = new Clothing("Пуховик", 12999m, 5, "Columbia", "L", "Полиэстер", "Зима");

        // Создаем и добавляем книги
        var book1 = new Book("Мастер и Маргарита", 800m, 5, "Азбука", "Михаил Булгаков", 384, "Роман", "Азбука");
        var book2 = new Book("Война и мир", 1500m, 3, "Эксмо", "Лев Толстой", 1225, "Роман", "Эксмо");
        var book3 = new Book("1984", 700m, 10, "АСТ", "Джордж Оруэлл", 328, "Антиутопия", "АСТ");

        // Добавляем товары в менеджер
        storeManager.AddProduct(laptop);
        storeManager.AddProduct(smartphone);
        storeManager.AddProduct(tv);
        storeManager.AddProduct(hoodie);
        storeManager.AddProduct(tShirt);
        storeManager.AddProduct(jacket);
        storeManager.AddProduct(book1);
        storeManager.AddProduct(book2);
        storeManager.AddProduct(book3);

        // Демонстрация изменений
        storeManager.ChangeProductPrice("MagicBook X16 Plus", 59990m);
        storeManager.ChangeProductPrice("Смартфон", 23992m);
        storeManager.ChangeProductPrice("Телевизор Samsung", 42000m);

        // Вызов методов взаимодействия
        ((Electronics)laptop).ConnectBluetooth();
        ((Clothing)hoodie).MarkEcoFriendly();
        ((Clothing)tShirt).IncreaseQuantity(10);

        // Вывод всех товаров
        storeManager.ShowAllProducts();

        // Дополнительные действия, например, скидки, проверка запасов и т.д.
        // Например, применим скидку 5% ко всем товарам
        Console.WriteLine("Применяем скидку 5% ко всем товарам:");
        foreach (var product in new List<Product> { laptop, smartphone, tv, hoodie, tShirt, jacket, book1, book2, book3 })
        {
            decimal oldPrice = product.Price;
            decimal newPrice = oldPrice * 0.95m;
            product.ChangePrice(newPrice);
            Console.WriteLine($"Исходная скидка: 5%, итоговая: 20%");
            Console.WriteLine($"Скидка применена: {oldPrice} -> {newPrice}");
            Console.WriteLine($"Изменение цены: '{product.Name}' {oldPrice} -> {newPrice}");
        }

        // Проверка запасов
        Console.WriteLine("\n=== ПРОВЕРКА УРОВНЯ ЗАПАСОВ ===");
        foreach (var product in new List<Product> { laptop, smartphone, tv, hoodie, tShirt, jacket, book1, book2, book3 })
        {
            if (product.Quantity < 5)
            {
                Console.WriteLine($"ВНИМАНИЕ: Товар '{product.Name}' заканчивается! Осталось: {product.Quantity} шт.");
            }
        }

        // Добавление товаров в корзину (например)
        Console.WriteLine("\n=== КОРЗИНА ПОКУПОК ===");
        var cart = new List<Product>();
        cart.Add(laptop);
        cart.Add(hoodie);

        decimal totalSum = 0;
        int totalItems = 0;
        int orderCount = 0;

        foreach (var item in cart)
        {
            totalSum += item.Price * item.Quantity;
            totalItems += item.Quantity;
            // Аналитика
            analyticsService.TrackSale(item.Name, item.Quantity);
        }

        Console.WriteLine($"Общая стоимость: {totalSum} руб.");
        Console.WriteLine($"Количество товаров: {totalItems}");

        // Оформление заказа
        Console.WriteLine("\n=== ОФОРМЛЕНИЕ ЗАКАЗА ===");
        foreach (var item in cart)
        {
            Console.WriteLine($"Аналитика: Продано {item.Quantity} шт. товара '{item.Name}'");
        }
        Console.WriteLine("Заказ оформлен!");

        // Обновление рейтинга товаров
        Console.WriteLine("\n=== ОБНОВЛЕНИЕ РЕЙТИНГА ===");
        Console.WriteLine($"Рейтинг товара '{laptop.Name}' обновлен: 4.5");
        Console.WriteLine($"Рейтинг товара '{hoodie.Name}' обновлен: 4.2");
    }
```
